```python
def write_merged_scenario_nc(file_df: pd.DataFrame, in_folder: Path(), tar_folder: Path):
    """
    Merge nc file by parameter, experiment, model
    Write the nc file to tar_folder
    """
    file_group = file_df.groupby(by=['scenario', 'param', 'model'])
    # return file_group
    count = 0
    for g in file_group.groups:
        try:
            df_ = file_group.get_group(g)
            tar_filename = '_'.join(g) + '.nc'
            if (tar_folder/tar_filename).exists():
                continue
            files_ser = str(in_folder) + '\\' + df_['filename']
            ds_set = [xr.open_dataset(file_i) for file_i in files_ser]
            count += 1
            print(count, tar_filename)
            xr.concat(ds_set, dim='time').to_netcdf(tar_folder/tar_filename)
        except Exception as e:
            failed_file = Path('../resources/merge_failed.txt')
            if not failed_file.exists():
                failed_file.write_text('')
            failed_file.write_text(f'{e}\n')
            print(e)
        finally:
            pass



cn_nc_folder = Path(r'F:\cy\Database\CMIP6\0_cn_nc')
files = [file for file in os.listdir(cn_nc_folder) if file.endswith('.nc')]
file_df = pd.DataFrame(files, columns=['filename'])
all_files = pd.read_csv('../resources/all_files.csv')
file_df = pd.merge(file_df, all_files, on='filename', how='inner')
write_merged_scenario_nc(file_df, cn_nc_folder, cn_nc_folder/'merged_nc')
```
